```python
# Importação das bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

# Criação de um exemplo de quadro Kanban
kanban_board = {
    'To Do': ['Tarefa 1', 'Tarefa 2', 'Tarefa 3'],
    'In Progress': ['Tarefa 4', 'Tarefa 5'],
    'Done': ['Tarefa 6', 'Tarefa 7', 'Tarefa 8']
}
kanban_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in kanban_board.items()]))
print(kanban_df)

# Visualização do Quadro Kanban
plt.figure(figsize=(10, 5))
sns.heatmap(kanban_df.isna(), cbar=False, cmap='Blues', linewidths=0.5, linecolor='black')
plt.title('Quadro Kanban')
plt.show()

# Interpretação: O quadro Kanban visualiza o fluxo de trabalho com três colunas: "To Do", "In Progress" e "Done". Isso ajuda a acompanhar o progresso das tarefas.

# Criação de um exemplo de Diagrama de Ishikawa (Espinha de Peixe)
G = nx.DiGraph()

# Adicionando nós principais
G.add_node('Problema de Produtividade')
causas = ['Pessoas', 'Processos', 'Tecnologia', 'Ambiente']
for causa in causas:
    G.add_edge('Problema de Produtividade', causa)

# Adicionando sub-causas
sub_causas = {
    'Pessoas': ['Falta de Treinamento', 'Desmotivação'],
    'Processos': ['Ineficiência', 'Falta de Padronização'],
    'Tecnologia': ['Ferramentas Obsoletas', 'Falta de Integração'],
    'Ambiente': ['Ruído', 'Iluminação']
}
for causa, sub in sub_causas.items():
    for s in sub:
        G.add_edge(causa, s)

# Visualização do Diagrama de Ishikawa
plt.figure(figsize=(14, 8))
pos = nx.spring_layout(G, k=1, iterations=20)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold', arrowsize=20)
plt.title('Diagrama de Ishikawa')
plt.show()

# Interpretação: O Diagrama de Ishikawa (Espinha de Peixe) identifica possíveis causas de um problema, agrupando-as em categorias. Isso facilita a análise e a solução de problemas.

# Criação de um exemplo de checklist 5S simplificado
five_s = {
    'Seiri': ['Eliminar itens desnecessários', 'Manter apenas o essencial'],
    'Seiton': ['Organizar itens de forma eficiente', 'Identificar locais de armazenamento'],
    'Seiso': ['Limpar regularmente', 'Manter padrões de limpeza'],
    'Seiketsu': ['Padronizar processos', 'Treinamento contínuo'],
    'Shitsuke': ['Manter disciplina', 'Fomentar cultura de melhoria contínua']
}
five_s_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in five_s.items()]))
print(five_s_df)

# Visualização do Checklist 5S
plt.figure(figsize=(10, 5))
sns.heatmap(five_s_df.isna(), cbar=False, cmap='Greens', linewidths=0.5, linecolor='black')
plt.title('Checklist 5S')
plt.show()

# Interpretação: A checklist 5S visualiza as etapas da metodologia 5S para manter um local de trabalho organizado e eficiente.

# Criação de um exemplo de gráfico de burndown para Scrum simplificado
sprint_days = list(range(1, 11))
tasks_remaining = [50, 45, 40, 35, 30, 25, 20, 15, 10, 5]

sprint_df = pd.DataFrame({
    'Dias': sprint_days,
    'Tarefas Restantes': tasks_remaining
})
print(sprint_df)

# Visualização do Gráfico de Burndown
plt.figure(figsize=(10, 5))
plt.plot(sprint_df['Dias'], sprint_df['Tarefas Restantes'], marker='o', linestyle='-', color='red')
plt.fill_between(sprint_df['Dias'], sprint_df['Tarefas Restantes'], color='red', alpha=0.1)
plt.title('Gráfico de Burndown do Sprint')
plt.xlabel('Dias')
plt.ylabel('Tarefas Restantes')
plt.grid(True)
plt.show()

# Interpretação: O gráfico de burndown acompanha o progresso de um sprint, mostrando quantas tarefas ainda restam a serem concluídas ao longo do tempo.

# Gráfico de Dispersão com Linha de Regressão simplificado para Scrum
plt.figure(figsize=(10, 5))
sns.regplot(x=sprint_df['Dias'], y=sprint_df['Tarefas Restantes'], ci=None, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Gráfico de Dispersão com Linha de Regressão')
plt.xlabel('Dias')
plt.ylabel('Tarefas Restantes')
plt.grid(True)
plt.show()

# Interpretação: O gráfico de dispersão com linha de regressão mostra a relação entre os dias do sprint e as tarefas restantes, indicando a tendência de conclusão das tarefas ao longo do tempo.
```

Quadro Kanban:

Este gráfico simplificado ajuda a visualizar o fluxo de trabalho dividindo as tarefas em três colunas: "A Fazer", "Em Andamento" e "Concluído". Isso permite acompanhar facilmente o progresso das tarefas e identificar em que estágio cada tarefa está.

Diagrama de Ishikawa:

O Diagrama de Ishikawa, também conhecido como Diagrama de Espinha de Peixe, identifica e categoriza as possíveis causas de um problema de produtividade. As causas principais (Pessoas, Processos, Tecnologia e Ambiente) são divididas em sub-causas específicas. Isso facilita a análise e a solução de problemas, permitindo identificar os pontos críticos que precisam de atenção.

Checklist 5S:

A checklist 5S simplificada visualiza as etapas da metodologia 5S para organizar e manter um local de trabalho eficiente. Cada uma das cinco colunas representa um dos 5S (Seiri, Seiton, Seiso, Seiketsu e Shitsuke) e inclui ações específicas para cada etapa. Isso ajuda a garantir que todas as etapas da metodologia sejam seguidas de forma consistente.

Gráfico de Burndown do Sprint:

Este gráfico simplificado mostra o progresso de um sprint de Scrum, indicando quantas tarefas restam a serem concluídas ao longo do tempo. A linha vermelha representa a quantidade de tarefas restantes em cada dia do sprint, ajudando a visualizar se o time está no caminho certo para completar todas as tarefas até o final do sprint. O gráfico facilita o monitoramento do progresso e a identificação de possíveis atrasos.

Gráfico de Dispersão com Linha de Regressão:

Este gráfico de dispersão mostra a relação entre os dias do sprint e as tarefas restantes, com uma linha de regressão indicando a tendência geral. Os pontos azuis representam as tarefas restantes em cada dia, enquanto a linha vermelha mostra a linha de regressão ajustada. A inclinação da linha de regressão ajuda a visualizar a tendência de conclusão das tarefas ao longo do sprint.
